```python
import pandas as pd

def AlignmentandMainArray(mainfile):
    df = pd.read_csv(mainfile)
    step_x = df['Value'][0]
    stepdist_x = df['Value'][1]
    step_y = df['Value'][2]
    stepdist_y = df['Value'][3]
    lkey_R = df['Value'][4]
    lkey_C = df['Value'][5]
    lkey_x = df['Value'][6]
    lkey_y = df['Value'][7]
    rkey_R = df['Value'][8]
    rkey_C = df['Value'][9]
    rkey_x = df['Value'][10]
    rkey_y = df['Value'][11]
    mask_x = df['Value'][12]
    mask_y = df['Value'][13]
    wafer_x = df['Value'][14]
    wafer_y = df['Value'][15]
    test_mask_x = df['Value'][16]
    test_mask_y = df['Value'][17]
    wafer_size = df['Value'][18]
    left_blade = df['Value'][19]
    right_blade = df['Value'][20]
    front_blade = df['Value'][21]
    rear_blade = df['Value'][21]

    print('Stepping:', '\n', 'X Step Distance:', stepdist_x / 1000, 'Columns:', step_x, '\n', 'Y Step Distance:', 
          stepdist_y / 1000, 'Rows:', step_y, '\n')
    if wafer_size == 200:
        print('Wafer diameter: 215')
    elif wafer_size == 150:
        print('Wafer diameter: 180')
    elif wafer_size == 100:
        print('\n', 'Wafer diameter: 160', '\n', 'Translate Origin: X: -18 , Y: 20', '\n')
    else:
        print('Unsupported wafer size')

    rkey_xoffset = round(((-1*(step_x/2-0.5)*stepdist_x+(step_x-rkey_C)*stepdist_x) - rkey_x) / 1000, 5)
    rkey_yoffset = round(((rkey_y - (-1*(step_y/2-0.5)*stepdist_y+(rkey_R-1)*stepdist_y) ) / 1000), 5)
    lkey_xoffset = round((((-1*(step_x/2-0.5)*stepdist_x+(step_x-lkey_C)*stepdist_x) - lkey_x) / 1000), 5)
    lkey_yoffset = round(((lkey_y - (-1*(step_y/2-0.5)*stepdist_y+(lkey_R-1)*stepdist_y)) / 1000), 5)
    print ('Alignment Keys:', '\n', 'Right:', '\n', 'R:', int(rkey_R), '\n', 'C:', int(rkey_C)
           , '\n', 'x:', rkey_xoffset, 'y:', rkey_yoffset, '\n', 'Left:', '\n', 'R:', int(lkey_R), '\n', 
           'C:', int(lkey_C), '\n', 'x:', lkey_xoffset, 'y:', lkey_yoffset, '\n')
    
    pshift_x = round((((-1*(step_x/2-0.5)*stepdist_x) + mask_x) - wafer_x) / 1000, 5)
    pshift_y = round((wafer_y - ((-1*(step_y/2-0.5)*stepdist_y) + mask_y) ) / 1000, 5)
    print('MAIN array:', '\n',  'Pass Shift:', '\n', 'X:', pshift_x, '\n', 'Y:', pshift_y, '\n')

    
    left_b = (left_blade * 5) / 1000 + 50
    right_b = 50 - (right_blade * 5) / 1000
    front_b = (front_blade * 5) / 1000 + 50
    rear_b = 50 - (rear_blade * 5) / 1000
    print('Aperture Blades:', '\n', 'Left:', left_b, '\n', 'Right:', right_b, '\n', 'Front:', front_b, '\n', 'Rear:', rear_b)
    
    return test_mask_x, test_mask_y, step_x, step_y, stepdist_x, stepdist_y
    
    
def Plug(plugfile, num_plugs):
    print('\n', '\n', 'Plug Passes:', '\n')
    df_p = pd.read_csv(plugfile)   
    left_blade = df_p['Plug Pass Reticle Blade Left Position'][0]
    right_blade = df_p['Plug Pass Reticle Blade Right Position'][0]
    front_blade = df_p['Plug Pass Reticle Blade Bottom Position'][0]
    rear_blade = df_p['Plug Pass Reticle Blade Top Position'][0]
    left_b = (left_blade * 5) / 1000 + 50
    right_b = 50 - (right_blade * 5) / 1000
    front_b = (front_blade * 5) / 1000 + 50
    rear_b = 50 - (rear_blade * 5) / 1000
    print('Aperture Blades:', '\n', 'Left:', left_b, '\n', 'Right:', right_b, '\n', 'Front:', front_b, '\n', 'Rear:', rear_b,)
    
    for i in range(num_plugs):
        x_offset = round((((-1*(step_x/2-0.5)*stepdist_x+(step_x-df_p['Plug Closest Column'][i])*stepdist_x) + test_mask_x) - 
        df_p['Bottom Left x on wafer'][i]) / 1000, 5)
        y_offset = round((df_p['Bottom Left y on wafer'][i] - ((-1*(step_y/2-0.5)*stepdist_y+
                                                               (df_p['Plug Closest Row '][i]-1)*stepdist_y)
                                                              + test_mask_y)) / 1000, 5)
        print('\n', 'Plug', str(df_p['Plug Number '][i]) + ':', '\n', 'R:', df_p['Plug Closest Row '][i], '\t', 'y:', y_offset, '\n',
              'C:', df_p['Plug Closest Column'][i], '\t', 'x:', x_offset)
    
    
        
def Map(mapfile, num_map_sites):
    print('\n', '\n', 'Map Passes:', '\n')
    df_m = pd.read_csv(mapfile)
    for i in range(num_map_sites):
        x_offset = round((((-1*(step_x/2-0.5)*stepdist_x+(step_x-df_m['Closest Column'][i])*stepdist_x) - 
                          df_m['Center of alignment site x on wafer'][i]) / 1000), 5)
        y_offset = round(((df_m['Center of alignment site y on wafer'][i] -
                           (-1*(step_y/2-0.5)*stepdist_y+(df_m['Closest Row'][i]-1)*stepdist_y) 
                           )) / 1000, 5)
        print('Plug', str(df_m['Map Site Number'][i]) + ':', '\n', 'R:', df_m['Closest Row'][i], '\t', 'y:', y_offset, '\n', 
             'C:', df_m['Closest Column'][i], '\t', 'x:', x_offset)
   
```

```python
test_mask_x, test_mask_y, step_x, step_y, stepdist_x, stepdist_y = AlignmentandMainArray('//mcrtp-file-01/users/brady.pearce/My Documents/Stepper/Stepper Job Alignment.csv')
Plug('//mcrtp-file-01/users/brady.pearce/My Documents/Stepper/Test Plug.csv', 6)
Map('//mcrtp-file-01/users/brady.pearce/My Documents/Stepper/Test Map.csv', 6)

```
